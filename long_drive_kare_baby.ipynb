```python
import requests
import time

from gtts import gTTS
from IPython.display import Audio, display

# ---------------------------
# CONFIG
# ---------------------------
API_KEY = "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"
URL = "https://api.groq.com/openai/v1/chat/completions"

headers = {
    "Authorization": f"Bearer {API_KEY}",
    "Content-Type": "application/json"
}

# conversation memory
messages = [
    {"role": "system", "content": "You are my crush on a long drive. Be fun, flirty, and caring."}
]

# ---------------------------
# TTS function
# ---------------------------
def speak_gtts(text, lang='en'):
    tts = gTTS(text=text, lang=lang, slow=False)
    tts.save("temp.mp3")
    display(Audio("temp.mp3", autoplay=True));


# ---------------------------
# Chat function
# ---------------------------
def chat(user_input):
    messages.append({"role": "user", "content": user_input})

    data = {
        "model": "openai/gpt-oss-120b",
        "messages": messages,
    }

    response = requests.post(URL, headers=headers, json=data)
    reply = response.json()

    if 'choices' in reply and len(reply['choices']) > 0:
        bot_reply = reply['choices'][0]['message']['content']
        messages.append({"role": "assistant", "content": bot_reply})
        print("Bot:", bot_reply)
        if bot_reply:  # Check if bot_reply is not empty before speaking
            speak_gtts(bot_reply)
    else:
        print("Error: API response did not contain 'choices'. Full response:")
        print(reply)


# ---------------------------
# Collision alert function
# ---------------------------
def check_distance(distance):
    """
    If distance is too small, trigger warning scream.
    """
    if distance < 10:  # threshold in cm (you can tune this)
        warning = "Aaaaaaahhh!! Watch out, we are about to collide!"
        print("Bot:", warning)
        return speak_gtts(warning)

# ---------------------------
# Main Loop
# ---------------------------
while True:
    user_msg = input("You (type 'quit' to exit, 'dist X' for distance): ")

    if user_msg.lower() in ["quit", "exit"]:
        break

    # distance input case
    if user_msg.lower().startswith("dist "):
        try:
            distance_val = int(user_msg.split()[1])
            check_distance(distance_val)
        except:
            print("âš ï¸ Invalid distance. Use: dist 15")
        continue



    # normal chatbot response
    chat(user_msg)
    time.sleep(5)
```
